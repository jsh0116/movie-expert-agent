# Life Coach Agent

OpenAI Agents SDK + Clean Architecture 기반 라이프 코치 에이전트

## Architecture
```
domain/          - 엔티티 (CoachingSession, CoachingMessage)
application/     - Use Cases, Context (유스케이스 조율)
infrastructure/  - 환경 설정 (RunConfig)
presentation/    - Agent, Tools 정의 (코칭 팁)
```

## 1. 환경 설정

````python
from infrastructure.config import load_run_config

run_config = load_run_config()
print("환경 설정 완료")
````

## 2. DI Composition Root - 의존성 조립

````python
from domain.entities import CoachingSession
from application.use_cases import CoachChatUseCase
from application.context import AppContext

session = CoachingSession()
context = AppContext(coach_chat=CoachChatUseCase(session))

print("의존성 조립 완료")
````

## 3. Agent 생성

````python
from presentation.agent import create_agent

agent = create_agent()
print(f"Agent: {agent.name}")
print(f"Tools: {[t.name for t in agent.tools]}")
````

## 4. 대화 실행

아래 셀을 실행하면 직접 질문을 입력할 수 있습니다.  
`input_items` 리스트가 대화 히스토리를 유지합니다.

````python
# 대화 히스토리 초기화
input_items = []
````

````python
from agents import Runner

user_input = input("질문을 입력하세요: ")

input_items.append({"role": "user", "content": user_input})
result = await Runner.run(agent, input_items, run_config=run_config)
input_items = result.to_input_list()

print(f"\nCoach: {result.final_output}")
````
