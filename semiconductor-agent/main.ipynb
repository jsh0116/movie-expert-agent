### 명령어
| 명령어 | 설명 |
|--------|------|
| `/인터뷰` | 모의 기술면접 시작 (LLM judge 40/30/30 루브릭) |
| `/qa [주제]` | 개념 학습 코치 — 소크라테스 방식 |
| `/진단` | 이해도 진단 + 도메인별 점수 시각화 |
| `quit` | 세션 종료 |

```python
%matplotlib inline

import sys
from dotenv import load_dotenv
load_dotenv()

from langchain_core.messages import HumanMessage
from semiconductor.adapters.graph import create_app

print("✅ 모듈 로드 완료")
```

```python
# ── 세션 설정 ────────────────────────────────────────────────────
COMPANY = "samsung_ds"   # "samsung_ds" | "sk_hynix"
DOMAIN  = None           # None = 모든 도메인, "소자" | "공정" | "회로" | "트렌드"
MAX_Q   = 5              # 모의면접 문제 수

app, state = create_app(company=COMPANY, domain=DOMAIN, max_questions=MAX_Q)
print(state["display_output"])
```

```python
# ── 대화 루프 ────────────────────────────────────────────────────
while True:
    try:
        user_input = input("\n>>> ").strip()
    except (EOFError, KeyboardInterrupt):
        print("\n세션을 종료합니다.")
        break

    if not user_input:
        continue
    if user_input.lower() in ("quit", "exit", "종료"):
        print("세션을 종료합니다.")
        break

    state["messages"] = [HumanMessage(content=user_input)]
    state = app.invoke(state)

    if state.get("display_output"):
        print("\n" + state["display_output"])
```
